```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('the_office_lines_scripts.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# 1. How many characters are there? What are their names?
num_characters = df['speaker'].nunique()
character_names = df['speaker'].unique()
print(f"Number of characters: {num_characters}")
print(f"Character names: {character_names}")

# 2. For each character, find out who has the most lines across all episodes
most_lines_per_character = df.groupby('speaker')['line_text'].count().idxmax()
print(f"Character with the most lines: {most_lines_per_character}")

# 3. What is the average of words per line for each character?
df['num_words'] = df['line_text'].apply(lambda x: len(str(x).split()))
avg_words_per_line = df.groupby('speaker')['num_words'].mean()
print("\nAverage words per line for each character:")
print(avg_words_per_line)

# 4. What is the most common word per character
most_common_word_per_character = df.groupby('speaker')['line_text'].apply(lambda x: pd.Series(str(x).split()).mode().iloc[0])
print("\nMost common word per character:")
print(most_common_word_per_character)

# 5. Number of episodes where the character does not have a line, for each character
episodes_without_lines = df[df['line_text'].isnull()].groupby('speaker')['episode'].nunique()
print("\nNumber of episodes where the character does not have a line:")
print(episodes_without_lines)

# 6. The average percent of lines each character contributed each episode per season
df['lines_count'] = df.groupby(['season', 'episode', 'speaker'])['line_text'].transform('count')
df['lines_percentage'] = df['lines_count'] / df.groupby(['season', 'episode'])['line_text'].transform('count') * 100
avg_lines_percentage = df.groupby('speaker')['lines_percentage'].mean()
print("\nThe average percent of lines each character contributed each episode per season:")
print(avg_lines_percentage)
```

```python
# Other Fun Office things

# 7. Example: Number of scenes each character appears in
num_scenes_per_character = df.groupby('speaker')['scene'].nunique()
print("\nNumber of scenes each character appears in:")
print(num_scenes_per_character)

# 8. Example: Average number of lines per episode for each character
avg_lines_per_episode = df.groupby(['speaker', 'episode'])['line_text'].count().groupby('speaker').mean()
print("\nAverage number of lines per episode for each character:")
print(avg_lines_per_episode)
```
